```python
import os
from google.colab import drive
drive.mount('/content/drive')
DATA_DIR = "/content/drive/MyDrive/CCPP/"

!ls $DATA_DIR
```

```python
train_file = 'Folds5x2_pp.xlsx'
train_fn = os.path.join(DATA_DIR, train_file)
print(train_fn)
```

```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_excel(train_fn)
```

```python
data_df.head()
```

>Define dependent and independent variable. ***Dependent variable*** in a target or the outcome where as the *indepenednt variables* are the features. **x** will hold all the features and **y** will hold the target

```python
x= data_df.drop(['PE'],axis=1).values
y= data_df['PE'].values
```

```python
print(x)
print(y)
```

>Split the dataset we have to trainig and test sets. In this case we set the test_size to be 0.3, which is 30% of the dataset will be trained by our model.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state=0)
```

>use linear regression training

```python
from sklearn.linear_model import LinearRegression
ml = LinearRegression()
ml.fit(x_train,y_train)
```

```python
y_pred = ml.predict(x_test)
print(y_pred)
```

```python
ml.predict([[14.96,41.76,1024.07,73.17]])
```

>Evaluate our model using r2 score

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

>plot results

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Results")
plt.ylabel("Predicted Results")
plt.title("Actual vs predicted")
```

>Print the predicted values of our model


```python
pred_y_df = pd.DataFrame({'Actual value':y_test,'predicted value':y_pred,'Difference': y_test-y_pred})
pred_y_df[0:20]
```
