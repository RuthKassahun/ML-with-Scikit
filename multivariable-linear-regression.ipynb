>Define the x(independent) and y (dependet or the target) variables. x includes all the features except the last column. y is our taget (the value we want to predict)

```python
x=realState.drop(['Y house price of unit area'],axis =1).values
y=realState['Y house price of unit area']
```

> For visualization purpose we can print our features and the target values

```python
print(x)
print(y)
```

>Now let's split our data into training data and test data, for this case we will assign the test sets to be 30% of the training set. I will implement this using sci-kit learn

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state =0)
```

>Let's implement multiple value linear regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

>Let's predict on the test sets

```python
y_pred = lr.predict(x_test)
print(y_pred)
```

>Let's compute r2score of our model

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

>let's compare the results with respect to predicted value

```python
pred_y_df=pd.DataFrame({'Actual value':y_test,'Predicted value':y_pred,'Diffrencen':y_test-y_pred})
pred_y_df[0:20]
```
